Write a program to calculate the entropy for each of the variables in the following dataset.

```python
# Import libraries
import numpy as np
import pandas as pd
```

```python
# Read the data file
df = pd.read_csv('/Users/riteshturlapaty/ai-ml-learning/AccelerateAI/7. DecisionTree/DailyQuiz/q5_data.csv')
df
```

```python
# Calculate entropy value
entropy_list=[]

for coln_name in df.columns:
    prob_val=df[coln_name].value_counts(normalize=True)
    prob_val_list=prob_val.to_list()
    prob_val_sum=np.sum(prob_val)
    n = [(i*np.log2(i)) for i in prob_val_list]
    entropy_list.append(round(-1 * sum(n),2))
```

```python
new_df=pd.DataFrame(list(zip(list(df.columns),entropy_list)),columns=['Column name','Entropy'])
new_df
```

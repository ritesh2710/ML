```python
# Lets convert A2 to float and A14 to int64
credit_data['A2'] = credit_data['A2'].astype('float')
credit_data['A14'] = credit_data['A14'].astype('int64')
```

```python
# Segregate column names as per their datatype i.e this is to convert the values into numeric
cat_var_list = credit_data[['A1', 'A4', 'A5', 'A6', 'A7', 'A9', 'A10', 'A12', 'A13']]
num_var_list = credit_data[['A2', 'A3', 'A8', 'A11', 'A14', 'A15']]
target_list=credit_data[['T']]
```

```python
# Create dummy values for columns where values are non-numeric
cat_var_dummies=pd.get_dummies(cat_var_list)
```

```python
# Convert target variable values into numeric. Map the values
target_list['T'] = target_list['T'].map({'+':1, '-':0})

# Check T
target_list['T'].value_counts()
```

```python
num_var_list.shape
```

```python
cat_var_dummies.shape
```

```python
new_credit_card = pd.concat([num_var_list,cat_var_dummies], axis=1)
```

```python
new_credit_card
```

```python
# Prepare X and y
X=new_credit_card
y=target_list
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.25, random_state=25)
```

```python
# Case 1) max_depth = 5, 6, and 7

case=['Case-1','Case-2','Case-3']
max_depth=[5,6,7]
sample_split=[10,15,20]
max_features=['sqrt','log2']
```

```python
param_list=[['Case-1','max_depth',5],
    ['Case-1','max_depth',6],
    ['Case-1','max_depth',7],
    ['Case-2','sample_split',10],
    ['Case-2','sample_split',15],
    ['Case-2','sample_split',20],
    ['Case-3','max_features','sqrt'],
    ['Case-3','max_features','log2']]
decision_tree_features=pd.DataFrame(param_list,columns=['Case', 'ParamName', 'ParamValue'])
```

```python
decision_tree_features
```

```python
# create a data frame, to store results
df_model_accuracy = pd.DataFrame(data={'case': [], 
                                 'tuning_parameter': [], 
                                 'value':[],
                                 'training_accuracy':[],
                                 'test_accuracy':[]
                                }
                          )
```

```python
for i in decision_tree_features.index:
    if decision_tree_features['Case'][i]=='Case-1':
        model_decision_tree = DecisionTreeClassifier(random_state=0, criterion='gini', max_depth = decision_tree_features['ParamValue'][i])
    elif decision_tree_features['Case'][i]=='Case-2':
        model_decision_tree = DecisionTreeClassifier(random_state=0, criterion='gini',min_samples_split = decision_tree_features['ParamValue'][i])
    elif decision_tree_features['Case'][i]=='Case-3':
        model_decision_tree = DecisionTreeClassifier(random_state=0, criterion='gini', max_features = decision_tree_features['ParamValue'][i])
    else:
        print('The parameter does not exist')
    model_decision_tree.fit(X, y)
    train_accuracy = np.round(model_decision_tree.score(X_train,y_train) * 100, 2)
    test_accuracy = np.round(model_decision_tree.score(X_test,y_test) * 100, 2)
    df_model_accuracy.loc[len(df_model_accuracy.index)] = [decision_tree_features['Case'][i], decision_tree_features['ParamName'][i], decision_tree_features['ParamValue'][i], train_accuracy, test_accuracy]
```
